**Python if statement**
- *simple decision-making statement*

Syntax:

if (condition):
    
           Executes this statement if condition is true
else:
    
          Executes this statement if condition is false

if_flow.PNG

Explanation:

In the flowchart, the diamond symbol indicates some condition that must be tested. In this
case, we are determining whether the condition Cold outside is true or false. If this condition is true, the action Wear a coat is performed. If the condition is false, the action is  
skipped. The action is conditionally executed because it is performed only when a certain
condition is true.

***Note: Each of these examples below demonstrates a basic use of comparison operators in Python. Students can change the hardcoded values to see how the output varies depending on the condition.***

**Example 1: Simple Comparison**

```python
num = 10

# Check if the number is greater than 5
if num > 5:
    print("The number is greater than 5.")
else:
    print("The number is 5 or less.")
```

**Example 2: Even or odd**

```python
num = 87

# Check if the number is even
if num / 2 == 0:
    print("The number is even.")
else:
    print("The number is odd.")
```

**Example 3: Check Within range (using <= less than or equal to operator)**

```python
temperature = 21

# Check if temperature is within 20-25 degree celsious
if 20 <= temperature <= 25:
    print("The temperature is comfortable.")
else:
    print("The temperature is outside the comfort zone.")
```

**Example 4: using == (Equality Operator)**

```python
favorite_fruit = "apple"

# Checking if the favorite fruit is apple
if favorite_fruit == "apple":
    print("Your favorite fruit is an apple.")
else:
    print("Your favorite fruit is not an apple.")
```

**Example 5: using != (Not Equal Operator)**

```python
weather = "sunny" #try changing sunny to rainy and observe the output!

# Checking if the weather is not rainy
if weather != "rainy":
    print("It is not raining. You might not need an umbrella.")
else:
    print("It is rainy. Don't forget your umbrella!")
```

**Example 6: using >= (Greater Than or Equal To Operator)**

```python
age = 16

# Checking if age is greater than or equal to 16
if age >= 16:
    print("You are eligible to apply for a driving permit.")
else:
    print("You are not yet eligible to apply for a driving permit.")
```
