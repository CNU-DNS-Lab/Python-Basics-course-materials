**Explanation:** As long as the test condition remains true, execute the statements from body of while loop.

**While Loop Examples**

**Example 1**

```python
# program to display numbers from 1 to 5

# initialize the variable
i = 1
n = 5

# while loop from i = 1 to 5
while i <= n:
    print(i)
    i = i + 1
```

**Example 2**

In the below example, the while iterates until the user enters zero. When the user enters zero, the test condition evaluates to False and the loop ends.

```python
# program to calculate the sum of numbers
# until the user enters zero

total = 0
number = int(input('Enter a number: '))

# add numbers until number is zero
while number != 0:
    total += number    # total = total + number

    # ask for integer input again
    number = int(input('Enter a number: '))


print('total =', total)
```

**Example 3: Infinite Loop**

If the condition of a loop is always True, the loop runs for infinite times (until the memory is full).

```python
age = 32

# the test condition is always True
while age > 18:
    print('You can vote')
```

**Example 4: while loop with else**

In Python, a while loop may have an optional else block.
In below example, the else part is executed after the condition of the loop evaluates to False.

```python
counter = 0

while counter < 3:
    print('Inside while loop')
    counter = counter + 1
else:
    print('Inside else block')
```

**Example 5: While loop with break**

```python
counter = 0

while counter < 3:
    # loop ends because of break
    # the else part is not executed
    if counter == 1:
        break
    print('Inside loop')
    counter = counter + 1
else:
    print('Inside else')
```

Scenario: Imagine you have a program that simulates a password login system. The user has three attempts to enter the correct password. If they enter the correct password within those three attempts, they gain access; otherwise, they are locked out.

```python
correct_password = '12345'
attempts = 3

while attempts > 0:
  user_input = input('enter the password:')

  if(user_input==correct_password):
    print('access granted. Welcome!')
    break
  else:
    attempts = attempts-1
    if attempts>0:
      print('wrong password, you have' ,attempts, 'attempts left.')
    else:
        print("Incorrect password. Locked out!")
```
