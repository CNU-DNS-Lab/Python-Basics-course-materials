**Boolean expressions**

 The expressions that are tested by the if statement are called **Boolean expressions**.

image.png

**Example 1**

you can practice this example in python interpreter - interactive mode too.(optional)

```python
x=1  #statement
```

```python
x<0  #write a boolean expression
```

```python
b = (x==0) #statement containing boolean expression
print(b)
```

**Example 2**

```python
age1 = 18;
is_adult = (age1 >= 18)  # This will be True if age is 18 or more, otherwise False

print(is_adult)
```

```python
age2 = 6;
is_adult = (age2 >= 18)  # This will be True if age is 18 or more, otherwise False

print(is_adult)
```
